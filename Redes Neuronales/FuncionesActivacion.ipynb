# Funciones de Activacion

Existen diversas funciones que realizan esta accion de activar o desactivar una neurona. Estas funciones lo que hacen es aplicarse dentro de la capa oculta, con el fin de filtrar ciertos imputs que no deseamos.

```python
# @hidden_cell
# se importan algunas funciones importantes
import math
import numpy as np
import matplotlib.pyplot as plt
# dominio de las graficas
x = np.linspace(-5, 3, 20)

# funciones comunes de activasion
sigmoide = lambda x: 1/(1 + np.exp(-x))
tangente = lambda x: (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
relu     = lambda x: np.maximum(0,x)

# Definimos algunos parmetros para la grfica
f = plt.figure(figsize=(15,10))
plt.title('Funciones de activacin')

plt.subplot(2, 2, 1)
plt.grid()
plt.plot(x, x,label="Identidad")
plt.legend()

plt.subplot(2, 2, 2)
plt.grid()
plt.plot(x, sigmoide(x),label="Sigmoide")
plt.legend()

plt.subplot(2, 2, 3)
plt.grid()
plt.plot(x, tangente(x),label="Tangente H.")
plt.legend()

plt.subplot(2, 2, 4)
plt.grid()
plt.plot(x, relu(x),label="Relu")
plt.legend()

# Mostramos la leyenda de las grficas
plt.show()

```
